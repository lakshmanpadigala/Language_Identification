```python
#this function makes word into constuient ngrams....
def ngrams(word):
    word=word.replace("\n", "")
    ngram=[]
    for i in range(len(word)):
        ngram.append(word[i])
    for i in range(len(word)-1):
        ngram.append(word[i]+word[i+1])
    for i in range(len(word)-2):
        ngram.append(word[i]+word[i+1]+word[i+2])
    for i in range(len(word)-3):
        ngram.append(word[i]+word[i+1]+word[i+2]+word[i+3])
    for i in range(len(word)-4):
        ngram.append(word[i]+word[i+1]+word[i+2]+word[i+3]+word[i+4])
    return ngram

def fileNgram(a,vocabulary):
  a1 = open(a,'r')
  b1 = open('b.txt','w')
  for line in a1:
    list2 = ngrams(line)
    for i in range(len(list2)):
      str1=str(list2[i])
      str1=str1+" "
      b1.write(str1)
    b1.write('\n')
  b1.close()
  a1.close()
  str1 = 'b.txt'
  return vocfun(str1,vocabulary)

def vocfun(b,vocabulary):
  b1 = open(b,'r')
  #vocabulary = {}
  for line in b1:
    for word in line.split():
      if word in vocabulary:
        vocabulary[word] += 1
      else: 
        vocabulary[word] = 1
  b1.close()
  return vocabulary
```

```python
def dicttodf(vocabulary,eng,b):
  df=[]
  b1 = open('b.txt','r')
  a1 = open('/content/eng2.txt','r')
  for line in b1:
    empty_voc={}
    list1=[]
    for word in vocabulary:
      empty_voc[word]=0
    for word in line.split():
      if word in empty_voc:
        empty_voc[word] = 1
    for count in empty_voc.values():
      list1.append(count)
    list1.append(1)
    #label 1 for language other than english...
    df.append(list1)
  for line in a1:
    empty_voc={}
    list1=[]
    for word in vocabulary:
      empty_voc[word]=0
    for word in line.split():
      if word in empty_voc:
        empty_voc[word] = 1
    for count in empty_voc.values():
      list1.append(count)
    list1.append(0)
    df.append(list1)
  a1.close()
  b1.close()
  col=[i for i in range(len(df[0])-1)]+['label']
  df1 = pd.DataFrame(df ,columns=col)
  return df1
  
```

```python
def binaryClassifier(eng,b):
  b1 = open(b,'r')
  a1 = open(eng,'r')
  vocabulary = {}
  for line in a1:
    for word in line.split():
      if word in vocabulary:
        vocabulary[word] += 1
      else: 
        vocabulary[word] = 1
  vocabulary = fileNgram(b,vocabulary)
  df1 = dicttodf(vocabulary,'/content/eng2.txt',b)
  print("dataframe shape:",df1.shape)
  #print(vocabulary)
  return vocabulary,df1
 
 
```

```python
import pandas as pd
lr_ben = LogisticRegression()
df_ben = pd.read_csv('/content/bengali.csv')  
x1 = df_ben.drop(['label'], axis=1)
y1 = df_ben.filter(['label'],axis=1)
kf = StratifiedKFold(n_splits=5)
lr_scores1 = []
lr_accuracy1 = []
for train_index,test_index in kf.split(x1,y1,groups=None):
  lr_ben.fit(x1.loc[train_index],y1.loc[train_index])
  y_pred=lr_ben.predict(x1.loc[test_index])
  lr_scores1.append(f1_score(y1.loc[test_index],y_pred))
  lr_accuracy1.append(accuracy_score(y1.loc[test_index], y_pred, normalize=True, sample_weight=None))

```

```python
print("bengali_f1=",np.mean(lr_scores1),"bengali_accc=",np.mean(lr_accuracy1))
```
